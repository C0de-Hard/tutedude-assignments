1. Creating a DF

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}
```

1. a) printing first 5 rows

```python
df = pd.DataFrame(data)
```

1. b) summary of df based on certain columns

```python
df[['Age','Salary']].describe()
#describe gives important statistical overview
```

1. c) Calculate the average salary of employees in the 'HR' department.  

```python
# df['Salary'].mean() considers all emp; without filter
df[df['Department'] == 'HR']['Salary'].mean()
```

2. Add a new column, 'Bonus'

```python
df['Bonus'] = 0.1 * df['Salary']
df
```

3. Filter the DataFrame to show employees aged between 25 and 30.

```python
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)] # multiple conds are
#written as (cond1) & (cond2) | (cond3) etc into bracket[]
filtered_df
```

4. Group the data by 'Department' and calculate the average salary for each department.  

```python
df.groupby('Department')['Salary'].mean()
```

5. Sort the DataFrame by 'Salary' in ascending order and save the result to a new CSV file.  

```python
new_df = df.sort_values(by='Salary')
new_csv = new_df.to_csv('new_df.csv', index=False)
new_df
```
