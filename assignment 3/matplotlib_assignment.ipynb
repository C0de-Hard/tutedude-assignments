```python
import matplotlib as mpl
```

1. Create a simple line plot for the following data:


```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
mpl.pyplot.plot(x, y)
#a done
mpl.pyplot.title('Line Plot')
mpl.pyplot.xlabel('X-axis')
mpl.pyplot.ylabel('Y-axis')
mpl.pyplot.grid(True)
#b done
```

2. Create a bar graph to represent the marks scored by students in a subject:  

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
mpl.pyplot.bar(students, marks, color='r')
#a done
mpl.pyplot.title('Bar Graph')
mpl.pyplot.grid(True)
#b done
```

3. Create a pie chart to represent the percentage distribution of a company’s revenue from different regions:  

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
# mpl.pyplot.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=90)
mpl.pyplot.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=90, explode=(0.1, 0, 0, 0))
'''please comment any one of the last 2 line'''
```

4. Generate a histogram to show the frequency distribution of randomly generated integers between 1 and 100 (sample size = 1000).  

```python
import numpy as np
#random dataset
data = np.random.randint(1, 101, 1000)
mpl.pyplot.hist(data, bins=50, color='pink', edgecolor='black')
#you can change the bin value
```
